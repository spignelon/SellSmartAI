## Fetching latest Instagram posts URLs from an insta username

```python
import requests
```

##### Define the API URL and headers

```python
API_URL = "https://instagram-scraper-api3.p.rapidapi.com/user_posts"
API_HOST = "instagram-scraper-api3.p.rapidapi.com"
```

##### API key and the username of the account and number of posts to fetch

```python
api_key = "f50cfb13fcmshb90a4e9e481c3a9p1cedbejsnb02f2e16e808"
username = input("Enter the Instagram username: ")

post_count = int(input("Enter the number of posts to fetch: "))
```

##### Fetch user posts from the Instagram API

```python
def fetch_user_posts(api_key, username, count=5):
    headers = {
        "x-rapidapi-host": API_HOST,
        "x-rapidapi-key": api_key,
    }
    params = {
        "username_or_id": username,
        "count": count,
    }

    try:
        response = requests.get(API_URL, headers=headers, params=params)
        response.raise_for_status()
        return response.json()  # Return the JSON response if successful
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None
```

```python
api_response = fetch_user_posts(api_key, username, post_count)

print("API Response:", api_response)
```

##### Process the API response to extract post URLs

```python
def extract_post_urls(api_response):
    if not api_response:
        return []

    try:
        all_data = api_response["data"]
        username = all_data["user"]["username"]
        all_posts_data = all_data["items"]
        posts = []

        for post in all_posts_data:
            post_code = post["code"]
            posts.append(f"https://www.instagram.com/{username}/p/{post_code}/")

        return posts
    except KeyError as e:
        print(f"KeyError in response: {e}")
        return []
```

```python
post_urls = extract_post_urls(api_response)

print("Extracted Post URLs:", post_urls)
```

##### Display the list of post URLs

```python
if post_urls:
    print(f"\nSuccessfully fetched {len(post_urls)} post URLs:")
    for i, url in enumerate(post_urls, 1):
        print(f"{i}: {url}")
else:
    print("No posts could be fetched.")
```
